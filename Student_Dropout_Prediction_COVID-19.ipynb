```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
path = "/content/drive/MyDrive/7th_Semester/Ai_Project/03.Dataset_Student_Dropout_Prediction_COVID-19.xlsx"
df=pd.read_excel(path)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df=df.drop(['your_openion',
            'help_for_continuing_study',
            'reson_for_stopped_studying',],axis=1)
```

```python
df.isnull().sum()
```

```python
df['problem_if_study_running'] = df['problem_if_study_running'].fillna(df['problem_if_study_running'].mode()[0])
```

```python
df['support_online'] = df['support_online'].fillna(df['support_online'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df['present_condition']=LE.fit_transform(df['present_condition'])
```

```python
df['present_condition'].value_counts()
```

```python
df.columns
```

```python
dummies=pd.get_dummies(df[['gender',
                           'current_area',
                           'current_institute',
                           'problem_if_study_running',
                           'support_online']])
```

```python
df=pd.concat([df,dummies],axis=1)
```

```python
df=df.drop(['gender',
            'current_area',
            'current_institute',
            'problem_if_study_running',
            'support_online' ],axis=1)
```

```python
df
```

```python
X=df.drop(['present_condition'],axis=1)
```

```python
Y=df['present_condition']
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto')
X_sm, y_sm = smote.fit_resample(X, Y)
y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(X_sm,y_sm,test_size=.20,random_state=42,stratify=y_sm)
```

```python
ytest.value_counts()
```

```python
ytrain.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier

DT=DecisionTreeClassifier()

DT.fit(xtrain,ytrain)
y_pred=DT.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest,y_pred))

print("\nAccuracy Score: {}".format(accuracy_score(ytest,y_pred)))

print("\nClassification Report:")
print(classification_report(ytest,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF=RandomForestClassifier()

RF.fit(xtrain,ytrain)
y_pred=RF.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest,y_pred))

print("\nAccuracy Score: {}".format(accuracy_score(ytest,y_pred)))

print("\nClassification Report:")
print(classification_report(ytest,y_pred))
```

```python
from sklearn.ensemble import  ExtraTreesClassifier

ETC =ExtraTreesClassifier()
ET=ETC

ET.fit(xtrain,ytrain)
y_pred=ET.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest,y_pred))

print("\nAccuracy Score: {}".format(accuracy_score(ytest,y_pred)))

print("\nClassification Report:")
print(classification_report(ytest,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ABC = AdaBoostClassifier()

ABC.fit(xtrain,ytrain)
y_pred=ABC.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest,y_pred))

print("\nAccuracy Score: {}".format(accuracy_score(ytest,y_pred)))

print("\nClassification Report:")
print(classification_report(ytest,y_pred))
```

```python
from sklearn.neural_network import MLPClassifier

MLPC = MLPClassifier()

MLPC.fit(xtrain,ytrain)
y_pred_gbc=MLPC.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest,y_pred_gbc))

print("\nAccuracy Score: {}".format(accuracy_score(ytest,y_pred_gbc)))

print("\nClassification Report:")
print(classification_report(ytest,y_pred_gbc))
```

```python
from sklearn.ensemble import BaggingClassifier

BC = BaggingClassifier(n_estimators = 500, max_samples = 0.5, max_features = 0.5)
BC.fit(xtrain,ytrain)
BC.score(xtest,ytest)
preddtc=BC.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest, preddtc))

print("\nAccuracy Score: {}".format(accuracy_score(ytest, preddtc)))

print("\nClassification Report:")
print(classification_report(ytest,preddtc))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(xtrain,ytrain)
preddtc=gnb.predict(xtest)

print("Confusion Matrix:")
print(confusion_matrix(ytest, preddtc))

print("\nAccuracy Score: {}".format(accuracy_score(ytest, preddtc)))

print("\nClassification Report:")
print(classification_report(ytest,preddtc))
```
